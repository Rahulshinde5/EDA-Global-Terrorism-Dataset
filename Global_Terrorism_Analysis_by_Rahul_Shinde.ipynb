### Variables Description 

*  **eventid (Event id):**  A unique identifier for each terrorist attack event in the dataset.

*  **Killed:** The number of people killed in the attack.
*  **Wounded:** The number of people wounded in the attack.
*  **Casualties:** The number of people killed or injured in the attack.
*  **Group name:** The name of the terrorist group responsible for the attack.

````python
# Chart - 5 visualization code
# We know the insights of attacks by overall year, and Region now further we study different scenarios by attacks on  Specific Region base

'''Total Casualties (Killed + Wounded) in each Region'''

casualties_R=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by= 'Casualty', ascending= False)
# casualties_R.head()



''' Number of people killed in each Region'''

killed_R=df[["Region","Killed"]].groupby("Region").sum().sort_values(by= 'Killed', ascending= False)
# killed_R.head()



'''Number of people wounded in each Region'''

wounded_R=df[["Region","Wounded"]].groupby("Region").sum().sort_values(by= 'Wounded', ascending= False)
# wounded_R.head()
````
